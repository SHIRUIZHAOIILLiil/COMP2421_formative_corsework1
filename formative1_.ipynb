```python
import mpmath
import csv
from matplotlib import pyplot as plt
```

### Implemenation


Use the required formula and implement it within a function. Users can choose the number of cycles themselves.

```python
def calculate_pi_achi(n):
    mpmath.mp.dps = 50
    p_sides = mpmath.mpf(1.0) / mpmath.sqrt(mpmath.mpf(3.0))
    for i in range(0, n):
        p_sides = (mpmath.sqrt(p_sides ** 2 + 1) - 1) / p_sides
    pi = 6 * 2 ** n * p_sides
    return pi
```

### Results


The result of the program will be saved to the data_pi.csv file, which includes the number of iterations n, polygon edges, calculated value of pi, absolute error, and relative error.

```python
if __name__ == '__main__':

    with open("data_pi.csv", 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Loop', 'Sides', 'pi', 'A_error', 'R_error'])

    for i in range(0, 51):
        outcome = calculate_pi_achi(i)
        absolute_error = mpmath.fabs(outcome - mpmath.pi)
        relative_error = mpmath.fabs(outcome - mpmath.pi) / mpmath.pi
        with open("data_pi.csv", 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([i, 6 * 2 ** i, outcome, absolute_error, relative_error])
```
